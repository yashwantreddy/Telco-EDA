```python
#Features/columns with categoricals -
cats = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 
        'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']
```

```python
onehots = pd.get_dummies(clean_df[cats], drop_first=True)
```

```python
final = pd.concat([clean_df, onehots], axis=1)
```

```python
feats = final.drop('Churn', axis=1)
```

```python
y = final.Churn_Yes
```

```python
clean_df.columns
```

```python
X = final.drop(['index', 'customerID', 'gender', 'SeniorCitizen', 'Partner',
       'Dependents', 'tenure', 'PhoneService', 'MultipleLines',
       'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
       'PaperlessBilling', 'PaymentMethod',
       'Churn', 'Churn_Yes'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import RobustScaler, MinMaxScaler, StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
preds = rf.predict(X_test)
```

```python
accuracy_score(y_test, preds)
```

```python
cm = confusion_matrix(y_test, preds)
```

```python
cm
```

```python
def plot_confusion_matrix(confusion_mat):    
    plt.clf()
    plt.imshow(confusion_mat, interpolation='nearest', cmap=plt.cm.Wistia)
    classNames = ['Negative Class','Positive Class']
    plt.title('Churn or no Churn')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    tick_marks = np.arange(len(classNames))
    plt.xticks(tick_marks, classNames, rotation=45)
    plt.yticks(tick_marks, classNames)
    s = [['TN','FP'], ['FN', 'TP']]
    for i in range(2):
        for j in range(2):
            plt.text(j,i, str(s[i][j])+" = "+str(confusion_mat[i][j]))
    plt.show()
```

```python
plot_confusion_matrix(cm)
```
